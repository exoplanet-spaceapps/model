# 06_report_export — 匯出互動 HTML（Plotly）

```python
import glob, os, json, pandas as pd
import plotly.express as px, plotly.io as pio
os.makedirs('docs', exist_ok=True)
csvs = sorted(glob.glob('outputs/candidates_*.csv'))
df = pd.read_csv(csvs[-1]) if csvs else pd.DataFrame({'bls_period_d':[], 'model_score':[]})
fig = px.scatter(df, x='bls_period_d', y='model_score', title='Period vs Score')
pio.write_html(fig, file='docs/index.html', full_html=True, include_plotlyjs='cdn')
print('Wrote docs/index.html')
```
